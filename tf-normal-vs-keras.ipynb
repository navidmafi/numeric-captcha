```python
class StandardizationLayer(keras.layers.Layer):
    def call(self, inputs):
        mean = tf.reduce_mean(inputs, axis=[1, 2], keepdims=True)
        stddev = tf.math.reduce_std(inputs, axis=[1, 2], keepdims=True)
        return (inputs - mean) / (stddev + 1e-6)  # Avoid division by zero
    
class PerImageStandardizationLayer(keras.layers.Layer):
    def call(self, inputs):
        return tf.image.per_image_standardization(inputs)
```
